# Chapter 1: Necessary Statistics - Part I: Discrete Probability

## Imports

```python
import random
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

## Uniform distribution

```python
%config InlineBackend.figure_format = 'retina'

dice_outcomes = ["1", "2", "3", "4", "5", "6"]

dice_outcomes_prob_ = [1/len(dice_outcomes) for item in dice_outcomes]

plt.bar(dice_outcomes, dice_outcomes_prob_)
plt.xticks(dice_outcomes);
plt.xlabel('Rolling dice outcome', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(np.arange(0, 0.48, step=0.16), fontsize=12)
plt.ylabel('Outcome probability', fontsize=15)
plt.title('PMF', fontsize=15)
plt.tight_layout()

plt.savefig("../plots/prob_discrete_uniform.png")
```

```python
num_of_tries = 10_000
num_of_occurrence_of_side_2 = 0

for roll in range(num_of_tries):
    
    outcome = np.random.choice(dice_outcomes) #Rolling the dice, which has a random outcome between 1 and 6
    
    if outcome == "2":
        num_of_occurrence_of_side_2 += 1
        
prob_outcome = round(num_of_occurrence_of_side_2/num_of_tries, 3)

print("Probability of occurrence of side 2 in 10_000 rolling trials: ", prob_outcome)
```

## Bernoulli Distribution

```python
exam_outcomes = ["pass", "fail"]

exam_outcomes_prob_ = [0.8, 1 - 0.8]

plt.bar(exam_outcomes, exam_outcomes_prob_)
plt.xticks(exam_outcomes);
plt.xlabel('Exam outcome', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(np.arange(0, 1.2, step=0.2), fontsize=12)
plt.ylabel('Outcome probability', fontsize=15)
plt.title('PMF', fontsize=15)
plt.tight_layout()

plt.savefig("../plots/prob_discrete_bernoulli.png")
```

## Binomial Distribution

```python
p_heads = .25
n_flips = 50

coin_distn = stats.binom(n_flips, p_heads)

possible_heads = list(range(0, 51))

plt.bar(possible_heads,
        coin_distn.pmf(possible_heads))
plt.savefig("../plots/prob_discrete_binomial.png")
```
